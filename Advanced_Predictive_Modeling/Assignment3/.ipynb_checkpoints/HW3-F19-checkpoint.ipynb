```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

class LinearRegression:
    
    def __init__(self, learning_rate, regularization, n_epoch):
        self.learning_rate = learning_rate
        self.n_epoch = n_epoch
        self.regularization = regularization
        self.coef = np.zeros(5) #one coef for each of the terms
        self.eps = 10**-6 #used only for adagrad
        self.cache = np.zeros(self.coef.shape[0]) #used only for adagrad and momentum
        self.mu = 0.9 #used in momentum
        self.X = X
        self.y = y
        
    def sgd(self, gradient):
        '''
        updates self.coef based on gradient using SGD
        '''
        for j in range(self.coef.shape[0]):
            self.coef[j] = self.coef[j] - self.learning_rate*gradient[j]
        return self.coef
        
    def adagrad(self, gradient):
        '''
        updates self.coef based on gradient using Adagrad
        '''
        for j in range(self.coef.shape[0]):
            self.cache[j] += gradient[j]**2
            self.coef[j] = self.coef[j] - ( (self.learning_rate*gradient[j])/(self.eps+self.cache[j])**0.5 )

        return self.coef
    
    def sgd_momentum(self, gradient):
        '''
        updates self.coef based on gradient using Sgd_momentum
        '''
        for j in range(self.coef.shape[0]):
            self.cache[j] = self.cache[j]*self.mu + self.learning_rate*gradient[j]            
            self.coef[j] = self.coef[j] - self.cache[j] 
        return self.coef 
        
    def fit(self, X, y, update_rule='sgd', plot=False): 
        '''
        Fits the model for the given X, y. It uses the update rule mentioned by the parameter update_rule
        and displays a plot of the coefficients vs epochs and mse vs epochs if plot is True. 
        
        -> use get_features to get the features from X
        -> for epoch in epochs:
            iterate through all x, y.
                compute prediction using linearPredict.
                compute gradient.
                pass this gradient to the respective update rule and update the coefficients
                keep track of mse and coefficients
        -> plot if required
        
        '''
        x_train = self.get_features(X)
        mse = np.zeros(self.n_epoch)
        weights = np.array([[0.0 for i in range(self.coef.shape[0])] for j in range(self.n_epoch)]) 
        
        for n in range(self.n_epoch):
            v = np.zeros(self.coef.shape[0])
            
            for i in range(x_train.shape[0]):
                yhat = self.linearPredict(x_train[i])
                gradient = np.zeros(self.coef.shape[0])
                
                for j in range(self.coef.shape[0]):
                    gradient[j] = 2*(yhat - y[i])*x_train[i][j] + 2*self.regularization*self.coef[j]
                
                if(update_rule=='sgd'):
                    self.coef = self.sgd(gradient)
                
                if(update_rule=='sgd_momentum'): 
                    self.coef = self.sgd_momentum(gradient)
                
                if(update_rule=='adagrad'):
                    self.coef = self.adagrad(gradient)
                                
            # PREDICT
            y_fit = self.linearPredict(x_train)
            
            # MSE
            for h in range(y_fit.shape[0]):
                mse[n] += (y[h]-y_fit[h])**2
            mse[n] = mse[n]/float(y_fit.shape[0])
            weights[n] = list(self.coef) 
        
        # PLOTS
        if(plot):
            f, (ax1,ax2) = plt.subplots(1,2, figsize = (15,6))
            ax1.plot([n+1 for n in range(self.n_epoch)], mse)
            title = 'Update rule:'+ update_rule + ' Learning Rate:'+ str(self.learning_rate)+ ' Reg:' + str(self.regularization)
            ax1.set_title(title)
            ax1.set_ylabel('mse')
            ax1.set_xlabel('epoch')
            
            for i in range(self.coef.shape[0]):
                ax2.plot([n+1 for n in range(self.n_epoch)],[row[i] for row in weights], label = 'w_'+str(i))
                title = 'Update rule:'+ update_rule + ' Learning Rate:'+ str(self.learning_rate)+ ' Reg:' + str(self.regularization)
                ax2.set_title(title)
                ax2.set_ylabel('mse')
                ax2.set_xlabel('epoch')
                ax2.legend(loc = 'center left',bbox_to_anchor = (1,0.5))
            return mse
        return mse,weights
        
    def get_features(self, X):
        '''
        X (input) is the nx2 dimensional array - n data points with X_1 and X_2.
        It returns a nX5 dimensional array - n data points with 1, X_1, X_1X_2, X_1^2X_2^2, X_2^3.
        '''
        x0 = np.ones((X.shape[0],1), dtype=int)
        x1 = np.array([row[0] for row in X])
        x1_x2 = np.array([row[0]*row[1] for row in X])
        x1_x2_2 = np.array([(row[0]**2)*(row[1]**2) for row in X])
        x2_3 =  np.array([row[1]**3 for row in X])
        return np.concatenate((x0,x1.reshape(-1,1),x1_x2.reshape(-1,1),x1_x2_2.reshape(-1,1),x2_3.reshape(-1,1)), axis=1)
        
    def linearPredict(self, X_features):
        '''
        returns the dot product of X and self.coef
        '''
        yhat = np.dot(X_features,self.coef)
        return yhat
        
```

```python
data = pd.read_csv('q1_samples.csv')
X = np.array([data['x1'].values, data['x2'].values]).T
y = data['y'].values
n_epochs = 15
learning_rate = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
regularization = [0, 10, 100]
```

```python
import itertools
def best_model(update_rule):
    mse_15 = []
    params_list = list(itertools.product(learning_rate, regularization))

    m = np.zeros(n_epochs)
    w = np.array([[0.0 for i in range(5)] for j in range(n_epochs)]) 

    for p in params_list:
        l, r = p
        lin = LinearRegression(learning_rate=l, regularization=r, n_epoch=n_epochs)
        m, w = lin.fit(X,y,update_rule=update_rule, plot=False)
        mse_15.append(m[n_epochs-1])

    mse_sorted = sorted(mse_15, reverse=False)
    params_list_new = []
    params_list_new.append(params_list[mse_15.index(mse_sorted[0])])
    params_list_new.append(params_list[mse_15.index(mse_sorted[1])])

    m = np.zeros(n_epochs)
    print('The top 2 hyperparameter combinations and MSE are:')
    for p in params_list_new:
        l, r = p
        lin = LinearRegression(learning_rate=l, regularization=r, n_epoch=n_epochs)
        m = lin.fit(X,y,update_rule=update_rule, plot=True)
        print('MSE for parameters of',l,'and',r,'is:',m[n_epochs-1])
```

### 3.
AdaGrad allowed for a larger initial setting of the learning rate (0.01 vs 0.001 for sgd and sgd_momentum). This is because AdaGrad adaptively scales the learning rate for each dimension. SGD uses the same learning rate at each iteration in all dimensions (no scaling). On the other hand, AdaGrad adaptively scaled the learning rate with respect to the accumulated squared gradient at each iteration in each dimension.

1. (5pts) In this question, we will learn to build a Decision Tree classifier. Load the q3_dataset.csv from the files. The column named 'label' is the $Y$ we are trying to predict using the features given in all other columns. As usual, we first split the data in a train and test set and build our model on train set and test the performance on the test set. Use the class tree.DecisionTreeClassifier (http://scikit-learn.org/stable/modules/classes.html#module-sklearn.tree) to fit a model with default parameters and report its performance on the test data. Now, visualize the tree, for visualisation, if your classifier object is called clf, use the given commands to save the generated tree as a '.dot' file using export_graphviz(https://scikit-learn.org/stable/modules/generated/sklearn.tree.export_graphviz.html) and paste the contents of the file in Webgraphviz (http://www.webgraphviz.com/) to generate the tree.


```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
## read data
data= pd.read_csv('q3_dataset.csv')
x = data.drop(['label'],axis=1)
y = data.label.values
```

```python
## create train/test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=20)
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score

clf = tree.DecisionTreeClassifier()
clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)
accuracy_score(y_pred,y_test)
```

```python
## save tree as .dot
with open("file_name.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f)
```

2. (10pts) We can guide the learning of a Decision Tree classifier by specifying various parameters governing its construction. For example, the max_depth parameter forces the algorithm to stop after at most that many levels, the min_samples_split parameter lower bounds the amount of data required to split nodes when learning and the parameter called criterion determines what function to use to create splits at each node. For both type of criteria, gini and entropy, test the max_depth values in the range 1, 2, ..., 20 , and min_samples_split in the range $2^{1}, 2^{2}, ..., 2^{15}$. Visualise the best classification trees that you get for different criterion parameter and observe how do they differ. 

```python
import itertools
max_depth_vals = [n+1 for n in range(20)]
min_samples_split_vals = [2**(n+1) for n in range(15)]
params = list(itertools.product(max_depth_vals, min_samples_split_vals))

best_params = {}
best_acc = {}

for c in ['gini','entropy']:
    acc = []
    for p in params:
        p_md, p_ms = p
        clf = tree.DecisionTreeClassifier(criterion=c, max_depth=p_md, min_samples_split=p_ms, random_state=20)
        clf.fit(x_train,y_train)

        y_pred = clf.predict(x_test)
        acc.append(accuracy_score(y_pred,y_test))
    best_params[c] = (params[acc.index(sorted(acc, reverse=True)[0])])
    best_acc[c] = max(acc)
    clf = tree.DecisionTreeClassifier(criterion=c, max_depth=best_params[c][0], min_samples_split=best_params[c][1], random_state=20)
    clf.fit(x_train,y_train)
    f = 'file_name_'+c+'.dot'
    with open(f, 'w') as f:
        f = tree.export_graphviz(clf, out_file=f)
```

##  Observation:
Gini and entropy are the measures of node impurity. These measures help in splitting the tree to maximize the information gain. In our data, gini criterion is performing better with lesser max_depth i.e. a shorter tree and higher min_samples_split. Gini results in a simpler model with comparable accuracy. Also, at the leaf nodes the model with gini has one at zero which is a pure node.

```python
best_params
best_acc

from IPython.display import Image
print('Tree for the best parameters for Gini:')
Image('tree_gini.png')
print('Tree for the best parameters for Entropy:')
Image('tree_entropy.png')
```
